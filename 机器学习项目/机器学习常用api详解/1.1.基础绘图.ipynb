1 初级绘图

1.1 绘制折线图

```python
import matplotlib.pyplot as plt
import random
from pylab import mpl#处理文字显示问题
mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
```

```python
#0.准备数据
x = range(60)
y = [random.uniform(15,18) for i in x]
#1.绘制画布
plt.figure(figsize=(20,8),dpi=100)
#2.绘制折线图
plt.plot(x,y)
#3.展示
```

1.2 实现标题刻度功能

```python
#0.准备数据
x = range(60)
y = [random.uniform(15,18) for i in x]
#1.绘制画布
plt.figure(figsize=(20,8),dpi=100)
#2.绘制折线图
plt.plot(x,y)
#2.1添加x，y刻度
x_ticks_label = ['11:{}'.format(i) for i in x]
y_ticks_label =range(40)
#2.2修改x，y轴坐标显示
plt.xticks(x[::5],x_ticks_label[::5])#坐标刻度不可以通过字符串进行更改
plt.yticks(y_ticks_label[::5])
#2.3添加网格
plt.grid(True,linestyle="--",alpha=0.5)#(开启虚线，网格格式，透明度)
#2.4添加标题
plt.xlabel('时间')
plt.ylabel('温度')
plt.title('中午11～12点温度变化',fontsize=20)#(标题，字体大小)
#2.5保存图片
plt.savefig("./text.png")#放在show前
#3.展示‘
plt.show()
```

1.3 绘制多个图像

```python
#0.准备数据
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(1,3) for i in x]##############################增加北京数据
#1.绘制画布
plt.figure(figsize=(20,8),dpi=100)
#2.绘制折线图
plt.plot(x,y_shanghai,label='上海')
plt.plot(x,y_beijing,color='r',linestyle='--',label='北京')#############################(<横数据>,<纵数据>，color颜色。linestyle线段样式，label图例标注)
#2.1添加x，y刻度
x_ticks_label = ['11:{}'.format(i) for i in x]
y_ticks_label =range(40)
#2.2修改x，y轴坐标显示
plt.xticks(x[::5],x_ticks_label[::5])#坐标刻度不可以通过字符串进行更改
plt.yticks(y_ticks_label[::5])
#2.3添加网格
plt.grid(True,linestyle="--",alpha=0.5)#(开启虚线，网格格式，透明度)
#2.4添加标题
plt.xlabel('时间')
plt.ylabel('温度')
plt.title('中午11～12点温度变化',fontsize=20)#(标题，字体大小)
#2.5显示图例
plt.legend(loc=0)####################################显示图例
#2.6保存图片
plt.savefig("./text.png")#放在show前
#3.展示‘
plt.show()
```

1.4 多个坐标系（面向对象）

```python
#0.准备数据
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(1,3) for i in x]#增加北京数据
#1.绘制画布
# plt.figure(figsize=(20,8),dpi=100)
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=100)##########################(行数，列数，尺寸，大小)
#2.绘制折线图
# plt.plot(x,y_shanghai,label='上海')
# plt.plot(x,y_beijing,color='r',linestyle='--',label='北京')#(<横数据>,<纵数据>，color颜色。linestyle线段样式，label图例标注)
axes[0].plot(x,y_shanghai,label='上海')
axes[1].plot(x,y_beijing,color='r',linestyle='--',label='北京')#(<横数据>,<纵数据>，color颜色。linestyle线段样式，label图例标注)
#2.1添加x，y刻度
x_ticks_label = ['11:{}'.format(i) for i in x]
y_ticks =range(40)
#2.2修改x，y轴坐标显示
# plt.xticks(x[::5],x_ticks_label[::5])#坐标刻度不可以通过字符串进行更改
# plt.yticks(y_ticks_label[::5])
axes[0].set_xticks(x[::5])
axes[0].set_yticks(y_ticks[::5])
axes[0].set_xticklabels(x_ticks_label[::5])
axes[1].set_xticks(x[::5])
axes[1].set_yticks(y_ticks[::5])
axes[1].set_xticklabels(x_ticks_label[::5])
#2.3添加网格
#plt.grid(True,linestyle="--",alpha=0.5)#(开启虚线，网格格式，透明度)
axes[0].grid(True,linestyle="--",alpha=0.5)#(开启虚线，网格格式，透明度)
axes[1].grid(True,linestyle="--",alpha=0.5)#(开启虚线，网格格式，透明度)
#2.4添加标题
# plt.xlabel('时间')
# plt.ylabel('温度')
# plt.title('中午11～12点温度变化',fontsize=20)#(标题，字体大小)
axes[0].set_xlabel('时间')
axes[0].set_xlabel('温度')
axes[0].set_title('中午11～12点温度变化',fontsize=20)
axes[1].set_xlabel('时间')
axes[1].set_xlabel('温度')
axes[1].set_title('中午11～12点温度变化',fontsize=20)
#2.5显示图例
# plt.legend(loc=0)#显示图例
axes[0].legend(loc=0)
axes[1].legend(loc=0)
#2.6保存图片
plt.savefig("./text.png")#放在show前
#3.展示‘
plt.show()
```
