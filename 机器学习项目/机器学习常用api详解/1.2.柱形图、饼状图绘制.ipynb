1 基础图形绘制

```python
import matplotlib.pyplot as plt
import random
from pylab import mpl#处理文字显示问题
mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
```

1.1 散点图绘制

```python
#0 准备数据
x = [225.98,246.9,634.9,345.8,234.8,635.8,98.3,647.8,863.5,245.9,356.8,412.4,536.6]
y = [225.98,246.9,634.9,345.8,234.8,635.8,98.3,647.8,863.5,245.9,356.8,412.4,536.6]
#1 创建画布
plt.figure(figsize=(20,8),dpi=100)
#2 绘制图像
plt.scatter(x,y)#########################
#3 图像显示
plt.show()
```

1.2 柱状图

```python
#0 准备数据
#0.1 电影名字
movie_name = ['雷神','诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','密战','狂暴巨兽','格斯拉2','狂暴绝地','绝地武士']
#0.2 横坐标
x = range(len(movie_name))
#0.3 票房数据
y = [77345,86543,22134,15654,14578,56843,11111,9932,85311,99999]
#1 创建画布
plt.figure(figsize=(10,4),dpi=400)
#2 绘制图像
plt.bar(x,y,color=['b','r','g','y','c','m','y','k','c','g'],width=0.7)
#2.1 修改x显示
plt.xticks(x,movie_name,size=8)
#2.2 添加网格
plt.grid(linestyle='--',alpha=0.8)
#2.3 添加标题
plt.title('电影收入人群对比')
#3 显示图像
plt.show()
```
