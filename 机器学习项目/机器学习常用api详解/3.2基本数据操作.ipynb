### 累计统计函数

```python
data = pd.read_csv('/Users/zhx/Downloads/day03-资料/2.code/data/stock_day.csv')#读取csv文件
data = data.sort_index()
data
```

```python
stock_rise = data['p_change']
stock_rise
```

```python
import matplotlib.pyplot as plt#绘图
```

```python
stock_rise.cumsum().plot()#consum表示累计求和 cumax累计求最大值 cumin累计求最小值 cuprod累计求积
plt.show()
```
