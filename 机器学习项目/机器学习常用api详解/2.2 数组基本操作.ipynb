## 1.3 生成固定范围数组

```python
np.linspace(0,100,11)#(开始数，停止数，生成数量)生成等差数列
```

```python
np.arange(10,100,2)#(开始数，停止数，间隔)生成等差数列
```

```python
np.logspace(0,2,3,base=2)#（开始，结束，数量）从开始到结束生成对应数量的base^x的等比数列
```

## 1.4 生成随即数组

### 1.4.1 正态分布

```python
x1 = np.random.normal(1.75,1,100000000)#按照正态分布（1.75，1）生成100000个随即数的数组
x1
```

```python
#1 创建画布
plt.figure(figsize=(20,8),dpi=100)
#2 绘制图像
plt.hist(x1,1000)#将数据的总区间分成1000等份，用高度表示该区间的概率（直方图）
#3 显示图像
plt.show()
```

```python
x2 = np.random.normal(0,1,[4,5])
x2
```

```python
#1 创建画布
plt.figure(figsize=(20,8),dpi=100)
#2 绘制图像
plt.hist(x2,20)#将数据分成20等份，用高度表示该区间的概率（直方图）
#3 显示图像
plt.show()
```

### 1.4.2 均匀分布

```python
x3 = np.random.uniform(-1,1,10000000)
x3
```

```python
#1 创建画布
plt.figure(figsize=(20,8),dpi=100)
#2 绘制图像
plt.hist(x3,1000)#将数据的总区间分成1000等份，用高度表示该区间的概率（直方图）
#3 显示图像
plt.show()
```
