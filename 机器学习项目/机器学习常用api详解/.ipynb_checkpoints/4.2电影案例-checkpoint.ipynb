# Metascore的分布情况
## 解决值方图对不齐刻度问题

```python
#绘制rating
movie['Metascore'].plot(kind = 'hist')#pandas绘制情况不好
```

```python
#1.创建画布
plt.figure(figsize = [20,8],dpi = 100)
#2.创建画布
plt.hist(movie['Metascore'].values,bins = 20)
#2.1添加刻度
max = movie['Metascore'].max()
min = movie['Metascore'].min()

t1 = np.linspace(min,max,num=21)#20组要用21条线来分
plt.xticks(t1)
#2.2添加网格
plt.grid()
#3.显示图像
plt.show()
```
