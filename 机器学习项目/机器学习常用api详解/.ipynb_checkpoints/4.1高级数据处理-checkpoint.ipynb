# 缺失值处理 缺失值为NaN
## 缺失替换

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movie = pd.read_csv('~/文档/read/IMDB-Movie-Data.csv')
```

```python
movie.head()
```

```python
pd.notnull(movie)#将NaN空值变为False
```

```python
np.all(pd.notnull(movie))#里面有缺失则返回False
```

```python
pd.isnull(movie)#有缺失返回True
```

```python
np.any(pd.isnull(movie))#有缺失则返回true
```

## 缺失删除

```python
movie.dropna()#存在nan的正行删除
```

```python
np.all(movie.dropna())#是否有缺失值
```

##  缺失写入

```python
average = movie['Revenue (Millions)'].mean()#求Revenue (Millions)的平均值
```

```python
movie['Revenue (Millions)'].fillna(average , inplace = True)#average 表示替换值 inplace表示是否要对原来的缺失值替换（默认false）
movie
```

```python
for i in movie.columns:
    if np.any(pd.isnull(movie[i])) == True:
        movie[i].fillna(movie[i].mean(), inplace = True)#循环处理每一列
```

```python
np.any(pd.isnull(movie)) #没有缺失值
```

# 数据离散化处理

```python
data = pd.read_csv('~/文档/read/stock_day.csv')
```

```python
data.head()
```

```python
p_change = data['p_change']#取p_change这一列
```

```python
#分成差不多数量的几组
qcut = pd.qcut(p_change , 10)
qcut
```

```python
#读取每段数据的数量
qcut.value_counts()
```

```python
#指定分组区间
bins  = [-100, -7, -5, -3, 0, 3, 5, 7, 100]#划定分组区间
pcut = pd.cut(p_change, bins)
pcut
```

```python
pcut.value_counts()
```

```python
dumies = pd.get_dummies(pcut, prefix = 'Rise')#传入参数（分组后的.CUT , columns标题名）
dumies
```

# 数据合并
## 无相同列索引的合并

```python
pd.concat([data,dumies], axis = 1)#axis 以行索引为基准合并列索引
```

## 存在相同列索引的合并

```python
left = pd.DataFrame({'key1': ['K0','K0','K1','K2'], 'key2': ['K0','K1','K0','K1'], 'A': ['A0','A1','A2','A3'], 'B': ['B0','B1','B2','B3']})
right = pd.DataFrame({'key1': ['K0','K1','K1','K2'], 'key2': ['K0','K0','K0','K0'], 'C': ['C0','C1','C2','C3'], 'D': ['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right,on =['key1','key2'])#默认内连接
```

```python
pd.merge(left, right,on =['key1','key2'],how = 'inner')#内连接不考虑有不一样key的行
```

```python
pd.merge(left, right,on =['key1','key2'],how = 'left')#左连接以左边为基准合并
```

```python
pd.merge(left, right,on =['key1','key2'],how = 'right')#左连接以右边为基准合并
```

```python
pd.merge(left, right,on =['key1','key2'],how = 'outer')#左右连接的合并
```

# 交叉表与透视表
## 交叉表

```python
data
```

```python
data.index
```

```python
time = pd.to_datetime(data.index)#转换数据为日期类型
time
```

```python
time.weekday#日期格式下的特殊函数
```

```python
time.day
```

```python
data['week'] = time.weekday
```

```python
data.head()
```

```python
data['p_n'] = np.where(data['p_change']>0,1,0)#将正负变为1，0
```

```python
data.head()
```

```python
count = pd.crosstab(data['week'],data['p_n'])#第一个参数的列表值作为行索引 第二个参数的列表值作为列索引 统计出所有数据对应的关系
count
```

```python
sum = count.sum(axis = 1).astype(np.float32)
sum
```

```python
ret = count.div(sum,axis = 0)
ret
```

```python
ret.plot(kind = 'bar',stacked = True)#stacked是堆积参数，即是否和并柱状图
plt.show()
```

## 透视表

```python
data.pivot_table(['p_n'],index ='week')#计算百分占比
```

# 分组与聚合

```python
col = pd.DataFrame({'color': ['white','red','green','red','green'],'object': ['pen','pencil','pencil','ashtray','pen'],'price1':[5.56,4.20,1.30,0.56,2.75],'price2':[4.75,4.12,1.60,0.75,3.15]})
```

```python
col
```

```python
col.groupby(['color'])['price1'].mean()#按照颜色分组 分组后的值是每一类的平均值  分组+聚合 DateFrame方法
```

```python
col['price1'].groupby(col['color']).mean()# 聚合+分组 Series方法
```

```python
col.groupby(['color'],as_index = False)['price1'].mean()#取消color为行索引
```
